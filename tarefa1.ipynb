```python
# prompt: Construa um modelo de regressão logística para prever a presença de doença cardíaca (alvo) usando variáveis ​​como idade, sexo, cp, chol, thalach, e ca.

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Assuming your dataframe is named 'df'
# Select the features and target variable
X = df[['age', 'sex', 'cp', 'chol', 'thalach', 'ca']]
y = df['target']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Create a Logistic Regression model
model = LogisticRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# prompt: Interpretar os coeficientes do modelo de regressão logística, focando nas variáveis ​​significativas. Discuta como essas variáveis ​​influenciam a probabilidade de presença de doença cardíaca.

# Get the coefficients and feature names
coefficients = model.coef_[0]
feature_names = X.columns

# Create a DataFrame to store the coefficients and feature names
coefficients_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})

# Sort the DataFrame by the absolute value of the coefficients to see the most influential features
coefficients_df['Abs_Coefficient'] = coefficients_df['Coefficient'].abs()
coefficients_df = coefficients_df.sort_values('Abs_Coefficient', ascending=False)

# Print the sorted coefficients
coefficients_df

# Interpretation:
# 1. Positive coefficients:
#   - Indicate that an increase in the corresponding feature increases the log-odds of having heart disease.
#   - For example, if the coefficient for 'age' is positive, it means that as age increases, the likelihood of having heart disease also increases (assuming other factors are constant).

# 2. Negative coefficients:
#   - Indicate that an increase in the corresponding feature decreases the log-odds of having heart disease.
#   - For example, if the coefficient for 'thalach' (maximum heart rate achieved) is negative, it means that as maximum heart rate increases, the likelihood of having heart disease decreases (assuming other factors are constant).

# 3. Magnitude of coefficients:
#   - The larger the absolute value of the coefficient, the stronger the influence of the feature on the prediction.

# Discussion:

# Based on the output of the coefficients_df, you can analyze which features are most significant in predicting heart disease.
# For example, if 'cp' (chest pain type) and 'ca' (number of major vessels colored by fluoroscopy) have large positive coefficients, this suggests that higher values of these features are associated with a higher likelihood of heart disease.
# Similarly, if 'thalach' has a large negative coefficient, this means that higher maximum heart rate is associated with a lower likelihood of heart disease.

# You can then discuss the clinical significance of these findings and how these variables might influence the probability of having heart disease.
# For example, if 'age' has a large positive coefficient, you can conclude that older individuals are at a higher risk of heart disease compared to younger individuals.
# If 'chol' (serum cholesterol) has a positive coefficient, you can discuss how higher cholesterol levels might increase the risk.
```

```python
# prompt: Calcule o AIC do modelo e discuta sua relevância na seleção do modelo.

import statsmodels.api as sm

# Assuming your model is named 'model' and your training data is X_train and y_train
X_train_with_constant = sm.add_constant(X_train)  # Add a constant term for the intercept
logit_model = sm.Logit(y_train, X_train_with_constant).fit()

# Calculate AIC
aic = logit_model.aic

print(f"AIC of the model: {aic}")

# Discussion of AIC's Relevance in Model Selection:

# The Akaike Information Criterion (AIC) is a measure of the relative quality of statistical models for a given set of data.
# It balances the goodness of fit of the model with the complexity of the model.
# A lower AIC value indicates a better model, meaning it provides a better balance between fit and complexity.

# In model selection, we often compare multiple models (e.g., with different features or different types of regression) using AIC.
# The model with the lowest AIC is generally considered the preferred model.

# In the context of your logistic regression model for heart disease prediction:
# 1. The calculated AIC value can be used to compare different logistic regression models built with different sets of features.
# 2. If you have another model with a lower AIC, it indicates that the alternative model might be a better fit for the data, potentially providing more accurate predictions while being more parsimonious (less complex).

# Limitations of AIC:
# - AIC is based on asymptotic properties and may not be accurate for small sample sizes.
# - It assumes that the model is correctly specified (i.e., the true underlying model is included in the set of candidate models).
```
